```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from countryinfo import CountryInfo

import matplotlib.pyplot as plt
import plotly.graph_objects as go

%matplotlib inline
```

```python
cases = pd.read_csv('data/confirm.csv')
recover = pd.read_csv('data/recover.csv')
deaths = pd.read_csv('data/death.csv')
```

```python
display(cases.head(), recover.head(), deaths.head())
```

```python
region = {}
subregion = {}

fails = []

regions = sorted(deaths['Country/Region'].unique())
for r in regions:
  try:
    country = CountryInfo(r)
    region[r] = country.region()
    subregion[r] = country.subregion()
  except:
    fails.append(r)
    print('No data for', r)
```

```python
region['Cabo Verde'] = 'Africa'
region['Comoros (the)'] = 'Africa'
region['Congo (Brazzaville)'] = 'Africa'
region['Congo (Kinshasa)'] = 'Africa'
region["Cote d'Ivoire"] = 'Africa'
region['Eswatini'] = 'Africa'
region['Gambia'] = 'Africa'
region['Sao Tome and Principe'] = 'Africa'
region['Sudan (the)'] = 'Africa'


subregion['Cabo Verde'] = 'Western Africa'
subregion['Comoros (the)'] = 'Southern Africa'
subregion['Congo (Brazzaville)'] = 'Middle Africa'
subregion['Congo (Kinshasa)'] = 'Middle Africa'
subregion["Cote d'Ivoire"] = 'Western Africa'
subregion['Eswatini'] = 'Southern Africa'
subregion['Gambia'] = 'Western Africa'
subregion['Sao Tome and Principe'] = 'Middle Africa'
subregion['Sudan (the)'] = 'Eastern Africa'
```

```python
def addColumns(df, region=region, subregion=subregion):
    # Adding in population as a column
    df['Region'] = df['Country/Region'].map(region)
    df['Subregion'] = df['Country/Region'].map(subregion)
    return df
```

```python
cases = addColumns(cases)
recover = addColumns(recover)
deaths = addColumns(deaths)
```

```python
# Re-arrange the data
cols = list(cases.columns)
cases = cases[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
recover = recover[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
deaths = deaths[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
```

```python
display(cases.head(), recover.head(), deaths.head())
```

```python
cases_africa = cases[cases['Region']=='Africa']
recover_africa = recover[recover['Region']=='Africa']
deaths_africa = deaths[deaths['Region']=='Africa']
```

```python
cases_africa['daily'] = cases_africa.iloc[:,-1] - cases_africa.iloc[:,-2]
recover_africa['daily'] = recover_africa.iloc[:,-1] - recover_africa.iloc[:,-2]
deaths_africa['daily'] = deaths_africa.iloc[:,-1] - deaths_africa.iloc[:,-2]
```

```python
display(cases_africa.head(), recover_africa.head(), deaths_africa.head())
```

```python
cols = list(cases_africa.columns)

cases_africa = cases_africa[['Country/Region' , 'Region' , 'Subregion'] + cols[-2:]]
recover_africa = recover_africa[['Country/Region' , 'Region' , 'Subregion'] + cols[-2:]]
deaths_africa = deaths_africa[['Country/Region' , 'Region' , 'Subregion'] + cols[-2:]]
```

```python
display(cases_africa.head(), recover_africa.head(), deaths_africa.head())
```

```python
cases_africa_sorted = cases_africa.sort_values(by='daily', ascending=False)
recover_africa_sorted = recover_africa.sort_values(by='daily', ascending=False)
deaths_africa_sorted = deaths_africa.sort_values(by='daily', ascending=False)
```

```python
display(cases_africa_sorted.head(), recover_africa_sorted.head(), deaths_africa_sorted.head())
```

```python
cases_africa_sorted.head().to_csv('top_cases.csv')
recover_africa_sorted.head().to_csv('top_recover.csv')
deaths_africa_sorted.head().to_csv('top_deaths.csv')
```

```python
cases_africa_sorted_sum = cases_africa_sorted.groupby('Region').sum()
recover_africa_sorted_sum = recover_africa_sorted.groupby('Region').sum()
deaths_africa_sorted_sum = deaths_africa_sorted.groupby('Region').sum()
```

```python
display(cases_africa_sorted_sum.head(), recover_africa_sorted_sum.head(), deaths_africa_sorted_sum.head())
```
